### 1. 선형대수란 무엇인가?
결론부터 말하자면 선형대수는 선형방정식을 풀기위한 방법론을 배우는 학문이다. 예를 들어 $2x + 3y = 0$과 같은 방정식이 있다면 우리의 목표는 우항($0$)과 같은 좌항들($x, y$)를 찾아내는 것이다. <br><br>

$3x + 1 = 7$과 같이 방정식이 하나이고, 미지수도 하나라면 우리는 이 $x$를 예측하기 매우 쉽다 ($x=2$) 그러나 만약 위의 $2x + 3y = 0$처럼 미지수가 두개라면 조건에 맞는 $x, y$를 찾는 것은 미지수가 하나 있을때 보다 훨씬 어려울 것이다. 우리의 직관등을 이용해 $x=-3, y=2$와 같은 값들의 집합을 찾을 수 있는데, 이것은 매우 비효율적이다. 그러나 만약 미지수가 2개, 방정식이 2개라면 `두 방정식의 관계를 이용하여 해를 찾아낼 수 있다.` 이 것이 바로 선형대수의 최종적인 목표이다. <br>

> Important: 선형대수의 목적 : `여러 방정식의 관계`를 이용해 방정식의 해를 찾는 것

<br><br>

### 2. "선형"이란 무엇인가
선형(Linear) 대수(Algebra)에서 선형이 갖는 의미는 무엇일까? 선형성이라는 것은 어떤 시스템(입/출력 존재)에 `입력값에 영향을 준 만큼 결과도 딱 그 만큼만 영향을 받는 성질`을 의미한다. 그러니까 입력 자체는 어떠한 변형이 되지 않고, 거기에 무언가를 더하거나 곱한다면 입력에 영향을 준 만큼만 출력이 영향을 받게 되는 성질이다. <br><br>

예를 들어 $y = 3x$과 같은 Linear한 시스템의 경우 기존 입력$x$에 대해 3이 곱해졌으므로 무조건 3이 곱해진 만큼만 영향을 받는다. 만약 입력이 $3$이였다면 출력은 $9$가 되고, 입력이 $25$였다면 출력은 $75$가 된다. 다른 예로 만약 $y = x + 4$와 같이 만들면, 입력이 $3$이였다면 출력은 $3+4$인 $7$이 되고, 입력이 $50$이였다면 출력은 $54$가 된다. 이 것이 선형적인 경우이다. <br><br>

$y = 2x^3$라는 Non-Linear한 시스템의 경우를 보면 그 값이 크게 변해버린다. $x = 8$라면 출력은 쌩뚱맞게 1024가 된다. (128배나 커져버린다) 선형성은 방정식들의 결과를 `예측 가능하게`만드는데, 이 성질은 방정식들의 관계를 이용해서 해를 구할 때 매우 중요한 성질이다.

<br><br>

### 3. "대수"란 무엇인가?
대수(Algebra)는 큰 수를 의미하는게 아니라 `어떤 다른 것이 수를 대신한다는 것`을 의미한다. 예기에서는 선형성을 가지는 방정식들이 수를 대신한다. 수의 본질은 기본적으로 연산에 있다. 우리가 흔히 "수"라고 부르는 정수, 유리수 등을 보면, 이러한 숫자들을 이용해서 연산이 가능하다. 선형대수에서는 선형방정식이 수를 대신하니, 선형 방정식들(벡터)이 기본 조각이 되고, 이 선형방정식(벡터)들을 서로 더하거나 빼는 등 연산을 한다. <br><br>

> 결과적으로 선형대수는 선형성을 가지는 방정식(벡터)들의 연산을 통해 방정식의 해를 구하기 위한 학문이다.
